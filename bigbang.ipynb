# Criando Formas e Animações Matemáticas

```python
import matplotlib.pyplot as plt
```

## Formas Básicas - 2D

Inicialmente ajustando o tamanho da figura para ficar na proporcao de um video do youtube

```python
def subplots_yt(show_grid=False, hide_axes=False, match_size=False):
    fig, ax = plt.subplots()

    fig.set_size_inches(19.2, 10.8)  # 1920x1080 pixels at 100 dpi
    ax.set_xlim(-9.6, 9.6)
    ax.set_ylim(-5.4, 5.4)
    if hide_axes:
        ax.axis('off')
    if show_grid:
        ax.grid(True)
    if match_size:
        fig.subplots_adjust(left=0, right=1, top=1, bottom=0)
        ax.set_position([0, 0, 1, 1])  # fill the whole figure

    return fig, ax
```

### Triângulo

```python
fig, ax = subplots_yt()

triangle = plt.Polygon([[-0.5, 0], [0.5, 0], [0, 1]], closed=True, edgecolor='black', facecolor='none')
ax.add_patch(triangle)

plt.show()
```

### Quadrado

```python
fig, ax = subplots_yt()

square = plt.Polygon([[0.5, 0.5], [-0.5, 0.5], [-0.5, -0.5], [0.5, -0.5]], closed=True, edgecolor='black', facecolor='none')
ax.add_patch(square)

fig.show()
```

## Texto

```python
fig, ax = subplots_yt()
ax.text(0, 0, r"$A = \pi r^2$", fontsize=60, ha='center', va='center')
fig.show()
```
